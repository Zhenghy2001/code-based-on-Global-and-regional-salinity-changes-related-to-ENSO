```python
import xarray as xr
import pandas as pd
```

```python
sada = xr.open_mfdataset('D:\ENSO_salinity\data\IAP_salinity\Salinity data orgin\*.nc', parallel=True)
salinity = sada.salinity
```

```python
FWC = 1 - salinity/1000
```

```python
FWC.name = 'freshwater'
```

```python
FWC.coords['time'] = pd.date_range('1980-1',periods=468,freq='M')
```

```python
FWC = FWC.groupby('time.month')
epflux_anomaly = FWC - FWC.mean(dim='time')
```
