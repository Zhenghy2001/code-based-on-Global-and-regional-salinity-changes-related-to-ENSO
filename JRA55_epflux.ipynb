```python
import xarray as xr
```

```python
eva = xr.open_mfdataset('D:/ENSO_salinity/data/epflux/JRA55/fcst_phy2m125.057_evp*')
pre = xr.open_mfdataset('D:/ENSO_salinity/data/epflux/JRA55/fcst_phy2m125.061_tprat*')
```

```python
eva = eva.EVP_GDS0_SFC_S130
pre = pre.TPRAT_GDS0_SFC_S130
```

```python
epflux = eva-pre
```

```python
epflux.name = 'ep'
```

```python
epflux.coords['g0_lat_1']
```

```python
epflux = epflux.to_dataset()
```

```python
epflux.coords['lon']=('g0_lon_2',pre.g0_lon_2)
epflux.coords['lat']=('g0_lat_1',pre.g0_lat_1)
epflux=epflux.swap_dims({'g0_lat_1':'lat','g0_lon_2':'lon'})
epflux=epflux.drop(['g0_lat_1','g0_lon_2'])
```

```python
epflux.coords['time']=('initial_time0_hours',pre.initial_time0_hours)
epflux=epflux.swap_dims({'initial_time0_hours':'time'})
epflux=epflux.drop(['initial_time0_hours'])
```

```python
epflux.to_netcdf('D:/ENSO_salinity/data/epflux/JRA55/epflux_jra55_mmday.nc')
```
