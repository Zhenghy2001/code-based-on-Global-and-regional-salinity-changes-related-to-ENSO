```python
import pandas as pd
import xarray as xr
```

```python
data = xr.open_mfdataset('D:\ENSO_salinity\data\Temperature\Temperature data orgin/*')
temp = data.temp
temp.coords['time'] = pd.date_range('1980-1',freq='M',periods=468)
temp = temp.groupby('time.month')
temp_anomaly = temp - temp.mean(dim='time')
temp_anomaly.to_netcdf('D:/ENSO_salinity/data/Temperature/IAP_temperature_anomaly.nc')
```

```python

data = xr.open_dataset('D:/ENSO_salinity/data/epflux/GPCP_OAflux/GPCP_OAflux_epflux_anomaly_mmday.nc')
epflux = data.epflux
s1 = epflux.isel(lon=range(0,36))
s2 = epflux.isel(lon=range(35,360))
s1.coords['lon'] = s1.coords['lon']+360
sada = xr.merge([s2,s1])
sada.to_netcdf('D:\ENSO_salinity\data\epflux\GPCP_OAflux\GPCP_OAflux_epflux_anomaly_mmday_new.nc')
```
