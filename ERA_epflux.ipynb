```python
import xarray as xr
```

```python
evap = xr.open_dataset('D:/ENSO_salinity/data/ERA5/Evaporation.nc')
precip = xr.open_dataset('D:/ENSO_salinity/data/ERA5/Total_precipitation.nc')
```

```python
evapr = evap.e*1000#mm/day
precip = precip.tp*1000#mm/day
```

```python
ep = evapr+precip
```

```python
ep.name= "epflux"
```

```python
ep = ep.to_dataset()
```

```python
ep.to_netcdf('D:/ENSO_salinity/data/epflux_era5_mmday.nc')
```
